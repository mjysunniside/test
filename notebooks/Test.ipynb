```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data')
```

```python
os.listdir(DATA_DIR)
```

```python
my_data = os.path.join(DATA_DIR, 'Inverters.csv')
df = pd.read_csv(my_data)
df.head(n=10)
```

```python
df2 = pd.read_csv(os.path.join(DATA_DIR, 'Storage.csv'))
df2.head(n=10)
```

```python
inverters_storage = pd.merge(df2, df, on='Address', how='outer', indicator=True)
inverters_storage.head(n=10)
```

```python
inverters_storage.to_csv('invertersStorage.csv')
```
